```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup



```

```python
# Set the URL you want to webscrape from
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

html = urlopen(url)

```

```python
soup = BeautifulSoup(html, 'lxml')
type (soup)
```

```python
#Get the title of the site
title = soup.title
print(title)
```

```python
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
#remove html tags
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean,'',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
#split the "0" column into multiple columns at the comma position
df1 = df[0].str.split(',', expand=True)
df1.head(20)
```

```python
#remove the "[" brackets from column 0
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
#remove the "\n" from column 2
df1[2] = df1[2].str.strip('\n]')
df1.head(10)

#remove the "]" from columne 0
df1[0] = df1[0].str.strip(']')
df1.head(10)
```

```python
#drop unwanted columns from df1
#df1.dropna(axis=1, inplace=True)
df1.head(10)
```

```python
#add table headers
col_labels = soup.find_all('th')
```

```python
#extract text in between html tags for table headers
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
#cleantext2 = cleantext2.strip('\n')
all_header.append(cleantext2)
print(all_header)
```

```python
#convert list of headers into a pandas df
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
#split column 0 into multiple columns at the common position for all rows
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
#df4 = df4.iloc[0].str.replace(" ","")
#print (df4)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
#drop rows with missing values
#df6 = df5.dropna(axis=0, how='any')
```

```python
#drop table header replicated as the first row
df7 = df5.drop(df5.index[0])
df7.head(20)
```

```python
#df7.rename(columns={'Borough':'Borough'}, inplace=True)
df7.rename(columns={'[Postcode':'PostalCode','Borough':'Borough',' Neighbourhood':'Neighbourhood'}, inplace=True)

#df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'}, inplace=True)

df7.head(20)
```

```python
#strip column names of whitespaces

df7.columns = df7.columns.str.strip()
```

```python
df8 = df7[['PostalCode','Borough','Neighbourhood']]
df8.head(20)
```

```python
#process only the borough's which don't have 'Not assigned'
df9 = df8.loc[df8['Borough'].str.strip() != 'Not assigned']

df9.head(100)
```

```python
#cell has Borough but not assigned Neighbourhood then Neighbourhood = Borough
df9.Neighbourhood.loc[(df9['Neighbourhood'].str.strip() == 'Not assigned')] = df9['Borough']
```

```python
df9.head(10)
```

```python
df10 = df9.drop_duplicates('PostalCode', keep='first').merge(df9.drop_duplicates('PostalCode', keep='last'),on='PostalCode')

df10.head(20)
```

```python
df11 = df10.assign(Neighbourhood = df10.Neighbourhood_x.astype(str) + ', ' + df10.Neighbourhood_y.astype(str))

df11.head(20)
```

```python
df12 = df11[['PostalCode','Borough_x','Neighbourhood']]

df12.rename(columns={'Borough_x':'Borough'}, inplace=True)

df12.head(30)
```
